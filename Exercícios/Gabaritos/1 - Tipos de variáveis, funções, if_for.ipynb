1) Crie uma função que recebe uma variável qualquer. Caso ela seja um número, imprima o resultado da raiz quadrada da soma dele com seu antecessor. Se for uma string, imprima uma f-string declarando qual a palavra inserida.

```python
def recebe_variavel(var):
  if type(var) == int:
    res = (2*var-1)**(0.5)
  else:
    res = f'A palavra é {var}.'
  return res
```

```python
print(recebe_variavel(5))
```

```python
print(recebe_variavel('leite'))
```

2) Crie uma função que imprima todas as letras de uma palavra digitada pelo usuário.

```python
def imprime_letras(palavra):
  for letra in palavra:
    print(letra)
```

```python
imprime_letras('banana')
```

3) Crie uma função que converta números em strings.

```python
def converte_numero(var):
  if type(var) == int or type(var) == float:
    res = str(var)
  else:
    res = var
  return res
```

```python
print(type(converte_numero(3.5)))
```

```python
print(type(converte_numero('banana')))
```

4) Crie uma função que, ao receber um número, retorne o valor da soma dele com seus quatro sucessores. Exemplo: se 5, 5+6+7+8+9.

```python
def soma_sucessores(var):
  if type(var) == int:
    seq = range(var,var+5)
    soma = 0
    for x in seq:
      soma = soma + x
    return soma
```

```python
print(soma_sucessores(5))
```
