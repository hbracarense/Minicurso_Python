1) Retorne ao exemplo do dicionário com as informações das pessoas que você conhece. Crie um dataframe distinto para cada um dos atributos, e faça merges para juntá-los em um único dataframe.

```python
import pandas as pd
from google.colab import drive, files
```

```python
df_idade = pd.DataFrame({'nome':('João', 'José', 'Luiz', 'Márcio', 'Maria', 'Felipe', 'Regina', 'Renata', 'Luciana', 'Vanessa'),
     'idade':(23,45,32,11,9,57,62,29,37,15)})

display(df_idade)
```

```python
df_altura = pd.DataFrame({'nome':('João', 'José', 'Luiz', 'Márcio', 'Maria', 'Felipe', 'Regina', 'Renata', 'Luciana', 'Vanessa'),
     'altura':(2.05,1.67,1.83,1.72,1.65,1.69,1.57,1.48,1.78,1.69)})

display(df_altura)
```

```python
df_peso = pd.DataFrame({'nome':('João', 'José', 'Luiz', 'Márcio', 'Maria', 'Felipe', 'Regina', 'Renata', 'Luciana', 'Vanessa'),
     'peso':(113,87.5,78,71,69,72.5,65,83,59,47)})

display(df_peso)
```

```python
df_cor_cabelo = pd.DataFrame({'nome':('João', 'José', 'Luiz', 'Márcio', 'Maria', 'Felipe', 'Regina', 'Renata', 'Luciana', 'Vanessa'),
     'cor_cabelo':('preto','preto','loiro','castanho', 'ruivo','preto','loiro','castanho','ruivo','preto')})

display(df_cor_cabelo)
```

```python
df = df_idade.merge(df_altura, left_on = 'nome', right_on = 'nome', how = 'left').merge(df_peso, left_on = 'nome', right_on = 'nome', how = 'left').merge(df_cor_cabelo, left_on = 'nome', right_on = 'nome', how = 'left')
display(df)
```
